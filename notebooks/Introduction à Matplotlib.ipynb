# Introduction à Matplotlib

[`matplotlib`](https://matplotlib.org) est une bibliothèque du langage de programmation Python destinée à tracer et visualiser des données sous forme de graphiques.
Elle est à la fois simple d'utilisation pour des usages basiques, mais également extrêmement personnalisable pour les personnes plus exigeantes sur l'apparence des figures générées.

### Ressources utiles

Les premières ressources à recommander sont évidemment la riche documentation de [`matplotlib`](https://matplotlib.org), notamment
le [guide d'utilisation](https://matplotlib.org/stable/users/index.html),
les [tutoriels](https://matplotlib.org/stable/tutorials/index.html),
les [types de graphiques](https://matplotlib.org/stable/plot_types/index.html#plot-types) ou encore
la [gallerie d'exemples](https://matplotlib.org/stable/gallery/index.html).

Concernant les ressources externes, une intéressante (avec malheureusement trop de publicitiés...) est le site internet [The Python Graph Gallery](https://python-graph-gallery.com), qui présente un grand nombre de type de graphiques et fournit le code ayant généré chacun des graphes.

### Autres paquets Python pour la visualisation de données

[`matplotlib`](https://matplotlib.org) étant un outil assez général pour générer des graphiques, d'autres paquets ont été développés en se focalisant davantage sur la visualisation en science des données.

Un autre paquet populaire pour créer des visualisations interactives pour des navigateurs internet est [`bokeh`](https://docs.bokeh.org).

Affichons la fonction carré évaluée sur les 5 premiers naturels (en commençant à partir de zéro) :

### 1.1 Personnalisation de la courbe

Le nombre d'arguments optionnels de la fonction [`matplotlib.pyplot.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html), permettant de personnaliser l'affichage de la courbe, est gigantesque (voir la classe [`matplotlib.lines.Line2D()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html)).
On va seulement évoquer les trois arguments suivants ici :

* `marker` : le style de balises (voir la [liste des balises disponibles](https://matplotlib.org/stable/api/markers_api.html)),
* `linestyle` : le style de lignes (voir la [liste des styles de lignes](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)),
* `color` : la couleur des lignes (et des balises par défaut) (voir la [liste des couleurs nommées](https://matplotlib.org/stable/gallery/color/named_colors.html)).

Voici un exemple de personnalisation de la courbe :

```python
plt.plot(
    x,
    x ** 2,
    marker='*',  # style de balises
    markersize=22,  # taille des balises
    markeredgecolor='C2',  # couleur des bords des balises
    markeredgewidth=1.5,  # épaisseur des bords des balises
    markerfacecolor='C4',  # couleur de la surface des balises
    linestyle='--',  # style de lignes reliant les points
    linewidth=2,  # épaisseur des lignes reliant les points
    color='C1',  # couleur des lignes reliant les points
)
plt.show()
```

On peut également afficher plusieurs courbes sur la même figure. Deux manières de faire sont possibles :
* *Première solution (la plus simple)* : il sufit de fournir les abscisses et les ordonnées de chacune des arguments comme premiers arguments de la fonction [`matplotlib.pyplot.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).
* *Deuxième solution (la plus flexible)* : il suffit d'appeler plusieurs fois la fonction [`matplotlib.pyplot.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

```python
plt.plot(x, x ** 2, x, 16 - x ** 2)
plt.show()
```

```python
plt.plot(
    x,
    x ** 2,
    marker='*',  # style de balises
    markersize=22,  # taille des balises
    markeredgecolor='C2',  # couleur des bords des balises
    markeredgewidth=1.5,  # épaisseur des bords des balises
    markerfacecolor='C4',  # couleur de la surface des balises
    linestyle='--',  # style de lignes reliant les points
    linewidth=2,  # épaisseur des lignes reliant les points
    color='C3',  # couleur des lignes reliant les points
)

plt.plot(
    x,
    16 - x ** 2,
    marker='P',  # style de balises
    markersize=12,  # taille des balises
    markeredgecolor='C6',  # couleur des bords des balises
    markeredgewidth=1.5,  # épaisseur des bords des balises
    markerfacecolor='C7',  # couleur de la surface des balises
    linestyle='-.',  # style de lignes reliant les points
    linewidth=1,  # épaisseur des lignes reliant les points
    color='C1',  # couleur des lignes reliant les points
)

plt.show()
```

## 1.2 Personnalisation de la figure

Il existe de nombreuses manières de personnaliser l'apparence d'une figure. La liste ci-dessous fournit les éléments les plus couramment utilisés :

* Pour initialiser (et personnaliser certains aspects de) la figure : [`matplotlib.pyplot.figure()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html)
* Pour récupérer ou changer les limites de l'axe des abscisses : [`matplotlib.pyplot.xlim()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html)
* Pour récupérer ou changer les limites de l'axe des ordonnées : [`matplotlib.pyplot.ylim()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html)
* Pour récupérer ou changer les marques sur l'axe des abscisses : [`matplotlib.pyplot.xticks()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)
* Pour récupérer ou changer les marques sur l'axe des abscisses : [`matplotlib.pyplot.yticks()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html)
* Pour ajouter un titre : [`matplotlib.pyplot.title()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html)
* Pour ajouter un label à l'axe des abscisses : [`matplotlib.pyplot.xlabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html)
* Pour ajouter un label à l'axe des ordonnées : [`matplotlib.pyplot.ylabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)
* Pour afficher la légende : [`matplotlib.pyplot.legend()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

Il est également possible de présenter tout le texte Matplotlib avec $\LaTeX$ en changeant la valeur de la clé `"text.usetex"` du dictionnaire `matplotlib.pyplot.rcParams` ou seulement certains textes de manière individuelle avec l'argument `usetex` (voir [Text rendering with LaTeX](https://matplotlib.org/stable/users/explain/text/usetex.html) et [Render math equations using TeX](https://matplotlib.org/stable/gallery/text_labels_and_annotations/tex_demo.html)).
**Il est nécessaire d'avoir une installation fonctionnelle de LaTeX sur sa machine pour utiliser cette fonctionnalité** (ce qui n'est pas le cas sur les machines virtuelles de l'ENSAI ou le SSP Cloud Datalab).

Enfin, il est possible de sauvegarder une figure avec la fonction [`matplotlib.pyplot.savefig()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html). En plus du nom du fichier (argument obligatoire), deux arguments optionnels pertinents sont :

* `dpi` (*dots per inch* ou [point par pource](https://fr.wikipedia.org/wiki/Point_par_pouce)) : il s'agit de la résolution de l'image ; plus le nombre est élevé, plus grande est la résolution (et la taille du fichier),

* `bbox_inches` : les images sauvegardées ont souvent pas mal de vide au niveau des bords ; une manière de réduire (si possible) ces bords vides est de fournir la valeur `"tight"` à cet argument.

Voici un exemple de personnalisation de la figure précédente :

```python
plt.figure(figsize=(10, 6))

plt.plot(
    x,
    x ** 2,
    marker='*',  # style de balises
    markersize=22,  # taille des balises
    markeredgecolor='C2',  # couleur des bords des balises
    markeredgewidth=1.5,  # épaisseur des bords des balises
    markerfacecolor='C4',  # couleur de la surface des balises
    linestyle='--',  # style de lignes reliant les points
    linewidth=2,  # épaisseur des lignes reliant les points
    color='C3',  # couleur des lignes reliant les points
    label=r'$x \mapsto x^2$',  # label pour la légende
)

plt.plot(
    x,
    16 - x ** 2,
    marker='P',  # style de balises
    markersize=12,  # taille des balises
    markeredgecolor='C6',  # couleur des bords des balises
    markeredgewidth=1.5,  # épaisseur des bords des balises
    markerfacecolor='C7',  # couleur de la surface des balises
    linestyle='-.',  # style de lignes reliant les points
    linewidth=1,  # épaisseur des lignes reliant les points
    color='C1',  # couleur des lignes reliant les points
    label=r'$16 - x \mapsto x^2$',  # label pour la légende
)

plt.xlim((-0.2, 4.2))  # limites de l'axe des abscisses
plt.ylim(((-1, 17)))  # limites de l'axe des ordonnées
plt.xticks(x, ['Zéro', 'Un', 'Deux', 'Trois', 'Quatre'])  # marques sur l'axe des abscisses
plt.yticks(np.arange(0, 17, 4))  # marques sur l'axe des ordonnées
plt.title(
    r"Courbes des fonctions $x \mapsto x^2$ et $x \mapsto 16 - x^2$", fontsize=18
)  # titre de la figure
plt.xlabel('Axe des abscisses', fontsize=12)  # label de l'axe des abscisses
plt.ylabel('Axe des ordonnées', fontsize=12)  # label de l'axe des ordonnées
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # légende
plt.savefig('Une jolie figure Matplotlib.png', dpi=400, bbox_inches='tight')  # sauvegarde de la figure

plt.show()
```

## 2. Les sous-figures

Il est courant de vouloir créer plusieurs sous-figures à l'intérieur. Pour ce faire, il y a deux approches possibles :

* *Première solution (la plus simple)* : utiliser la fonction [`matplotlib.pyplot.subplot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) ;
* *Deuxième solution (la plus flexible)* : utiliser [`matplotlib.pyplot.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html), qui renvoie une [`matplotlib.figure.Figure()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) et un tableau de [`matplotlib.axes.Axes()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html) ; chaque sous-figure a ses propres `Axes`.

Voici des illustrations d'utilisation de ces deux fonctions :

```python
plt.figure(figsize=(10, 4))

# Le premier argument est le nombre de lignes, le deuxième le nombre de colonnes et
# le troisième l'indice de la sous-figre (en commençant à compter à partir de 1)
plt.subplot(1, 2, 1)
plt.plot(x, x ** 2)
plt.title("Titre 1")
plt.xlabel("Axe des abscisses 1")
plt.ylabel("Axe des ordonnées 1")

plt.subplot(1, 2, 2)
plt.plot(x, 16 - x ** 2)
plt.title("Titre 2")
plt.xlabel("Axe des abscisses 2")
plt.ylabel("Axe des ordonnées 2")

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Ici, axes est un tableau NumPy en deux dimensions.
# Pour accéder à chaque axe, il suffit d'indexer ce tableau

# Figure en haut à gauche
axes[0, 0].plot(x, x ** 2, 'o-', color='C0')
axes[0, 0].set_title("Titre 1")
axes[0, 0].set_xlabel("Axe des abscisses 1")
axes[0, 0].set_ylabel("Axe des ordonnées 1")

# Figure en haut à droite
axes[0, 1].plot(x, 16 - x ** 2, '<-', color='C1')
axes[0, 1].set_title("Titre 2")
axes[0, 1].set_xlabel("Axe des abscisses 2")
axes[0, 1].set_ylabel("Axe des ordonnées 2")

# Figure en bas à gauche
axes[1, 0].plot(x, x ** 0.5, 'P-', color='C2')
axes[1, 0].set_title("Titre 3")
axes[1, 0].set_xlabel("Axe des abscisses 3")
axes[1, 0].set_ylabel("Axe des ordonnées 3")

# Figure en bas à droite
axes[1, 1].plot(x, x ** 3, 's-', color='C3')
axes[1, 1].set_title("Titre 4")
axes[1, 1].set_xlabel("Axe des abscisses 4")
axes[1, 1].set_ylabel("Axe des ordonnées 4")

# Pour ajuster automatiquement l'espace entre les sous-figures
fig.tight_layout()

plt.show()
```

```python
def single_barplot():
    # Abscisses
    x = np.arange(10)
    
    # Hauteurs
    height = np.random.randint(low=1, high=100, size=x.size)
    
    # Barplot
    plt.bar(x=x, height=height)
    
    # Ajout des hauteurs exactes au-dessus des barres
    for x_text, y_text in zip(x, height):
        plt.text(x=x_text, y=y_text, s=y_text, ha='center', va='bottom')

    # Barres de l'axe des abscisses
    plt.xticks(x)
    
    # Affichage de la figure
    plt.show()


single_barplot()
```

```python
def double_barplot():
    # Abscisses
    x = np.arange(10)
    
    # Hauteurs
    height_1 = np.random.randint(low=1, high=10, size=x.size)
    height_2 = np.random.randint(low=1, high=10, size=x.size)
    
    # Barplots
    plt.bar(x=x, height=height_1, width=-0.4, align='edge')
    plt.bar(x=x, height=height_2, width=+0.4, align='edge')
    
    # Barres de l'axe des abscisses
    plt.xticks(x)
    
    # Affichage de la figure
    plt.show()


double_barplot()
```

```python
def triple_barplot():
    # Abscisses
    x = np.arange(10)
    
    # Hauteurs
    height_1 = np.random.randint(low=1, high=10, size=x.size)
    height_2 = np.random.randint(low=1, high=10, size=x.size)
    height_3 = np.random.randint(low=1, high=10, size=x.size)
    
    # Barplots
    width = 0.2
    plt.bar(x=x-width, height=height_1, width=width, label='Échantillon 1')
    plt.bar(x=x, height=height_2, width=width, label='Échantillon 2')
    plt.bar(x=x+width, height=height_3, width=width, label='Échantillon 3')
    
    # Légende
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
    # Barres de l'axe des abscisses
    plt.xticks(x)

    # Affichage de la figure
    plt.show()


triple_barplot()
```

```python
def stacked_barplot():
    # Échantillon de taille 1 million
    size = 1_000_000
    
    # Différentes statistiques sur l'âge
    age_min = 18
    age_max = 68
    age_middle = (age_min + age_max) / 2
    
    # Discrétisation de l'âge pour le barplot
    age_step = 5
    age_bins = np.arange(age_min, age_max + age_step, age_step)

    # Simulation de l'âge selon une loi uniforme
    age = np.random.uniform(low=age_min, high=age_max, size=size)

    # Dépenses de types 1, 2 et 3 (en pourcentage du revenu)
    x1 = np.clip(
        22 + 20 * (age - age_min) / (age_max - age_min) + 2 * np.random.rand(size),
        0, 45
    )
    x2 = np.clip(
        26 + 23 * abs(age - age_middle) / (age_max - age_middle) + 3 * np.random.rand(size),
        0, 45
    )
    x3 = 100 - x1 - x2
    
    # Crée la matrice  ayant pour colonnes x1, x2 et x3
    X = np.c_[x1, x2, x3]

    # Calcule le pourcentage moyen pour chaque type de dépense pour chaque tranche d'âge
    X_binned = np.asarray([
        X[np.where((age > low) & (age <= high))].mean(axis=0)
        for low, high in zip(age_bins[:-1], age_bins[1:])
    ])

    # Barplots empilés
    bottom = np.zeros(age_bins.size - 1)
    for i in range(3):
        plt.bar(
            x=age_bins[:-1], height=X_binned[:, i], width=age_step, edgecolor='black',
            align='edge', bottom=bottom, label=f'Dépense {i + 1}'
        )
        bottom += X_binned[:, i]

    # Légende
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    # Axe des abscisses
    plt.xticks(age_bins)
    plt.xlabel('Âge')

    # Axe des ordonnes
    plt.ylabel('Pourcentage du revenu')
    
    # Limites
    plt.xlim((age_min, age_max))
    plt.ylim((0, 100))
    
    # Titre
    plt.title("Répartition des dépenses (en pourcentage du revenu)\npar tranche d'âge")


stacked_barplot()
```

```python
def horizontal_barplot():
    # Source : https://matplotlib.org/stable/gallery/lines_bars_and_markers/horizontal_barchart_distribution.html
    
    category_names = [
        'Strongly disagree', 'Disagree', 'Neither agree nor disagree',
        'Agree', 'Strongly agree'
    ]
    results = {
        'Question 1': [10, 15, 17, 32, 26],
        'Question 2': [26, 22, 29, 10, 13],
        'Question 3': [35, 37, 7, 2, 19],
        'Question 4': [32, 11, 9, 15, 33],
        'Question 5': [21, 29, 5, 5, 40],
        'Question 6': [8, 19, 5, 30, 38]
    }

    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)

    ax.legend(ncols=len(category_names), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')

    plt.show()
    
    
horizontal_barplot()
```

```python
def histogram_density():
    from scipy.stats import norm

    # Un million de réalisations indépendantes selon la loi normale centrée réduite
    data = np.random.randn(1_000_000)

    # Histogramme correspondant, normalisé en densité
    plt.hist(data, bins=100, density=True, label='Densité empirique')

    # Densité de la loi normale centrée réduite
    x = np.linspace(-4, 4, 1001)
    plt.plot(x, norm.pdf(x), label='Densité théorique')

    # Légende
    plt.legend()

    # Affichage de la figure
    plt.show()


histogram_density()
```

```python
def subplots():
    x = np.linspace(-5, 5, 1001)

    fig, axes = plt.subplots(2, 1, figsize=(6, 9))

    # Sigmoïde
    sigmoid = lambda x: 1 / (1 + np.exp(-x))
    axes[0].plot(x, sigmoid(x), zorder=1, label='Fonction')
    axes[0].plot(x, sigmoid(x) * (1 - sigmoid(x)), zorder=0, label='Dérivée')
    axes[0].set_title('Sigmoïde', fontsize=12);
    axes[0].legend(fontsize=10)

    x_min, x_max = axes[0].get_xlim()
    y_min, y_max = axes[0].get_ylim()
    axes[0].annotate(
        text='', xy=(x_max - 0.01, 0), xytext=(x_min, 0), zorder=-2,
        arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4)
    )
    axes[0].annotate(
        text='', xy=(0, y_max - 0.01), xytext=(0, y_min), zorder=-2,
        arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4)
    );

    # Tangente hyperbolique
    axes[1].plot(x, np.tanh(x), zorder=1, label='Fonction')
    axes[1].plot(x, 1 - np.tanh(x) ** 2, zorder=0, label='Dérivée')
    axes[1].set_title('Tangente hyperbolique', fontsize=12);
    axes[1].legend(fontsize=10)

    x_min, x_max = axes[1].get_xlim()
    y_min, y_max = axes[1].get_ylim()
    axes[1].annotate(
        text='', xy=(x_max - 0.01, 0), xytext=(x_min, 0), zorder=-2,
        arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4)
    )
    axes[1].annotate(
        text='', xy=(0, y_max - 0.01), xytext=(0, y_min), zorder=-2,
        arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4)
    );
    
    plt.show()

    
subplots()
```

```python
def imshow_contour():

    from matplotlib.colors import SymLogNorm

    def rosenbrock(x, y, epsilon=0.01):
        return (1 - x) ** 2 + (y - x ** 2) ** 2 + epsilon

    x = np.linspace(-2, 2, 1001)
    y = np.linspace(-1, 3, 1001)

    X, Y = np.meshgrid(x, y)

    Z = rosenbrock(X, Y)

    plt.imshow(Z, origin='lower', extent=[-2, 2, -1, 3], cmap='RdYlBu_r',
               norm=SymLogNorm(linthresh=0.0001, vmin=Z.min(), vmax=Z.max()))
    plt.colorbar()
    plt.contour(X, Y, Z, 10 ** np.arange(-1, 2, dtype='float64'))
    
    plt.title("Fonction de Rosenbrock", fontsize=14)

    plt.show()

    
imshow_contour()
```

```python
def linear_regression():

    from sklearn.linear_model import LinearRegression

    rng = np.random.RandomState(42)
    n_samples = 11
    x = np.linspace(-1, 2, n_samples)
    y = (0.7 * x) + 2 * rng.uniform(size=n_samples)
    y[3] += 0.5

    clf = LinearRegression().fit(x.reshape(n_samples, 1), y)
    y_pred = clf.predict(x.reshape(n_samples, 1))

    plt.figure(figsize=(8, 4))

    # Plot training data
    plt.scatter(x, y, s=20, label='Sortie')

    # Plot prediction
    plt.scatter(x, y_pred, color='C1', label='Prédiction', s=20)
    plt.plot(x, y_pred, color='C1')
    for i in range(n_samples):
        label = 'Erreur' if i == 0 else None
        plt.plot([x[i], x[i]], [y[i], y_pred[i]], color='C1', linestyle='--',
                 zorder=-5, label=label)

    # Labels
    plt.xlabel('$x$')
    plt.ylabel('$y$')

    # Legend
    handles, labels = plt.gca().get_legend_handles_labels()
    order = [1, 2, 0]
    plt.legend([handles[idx] for idx in order], [labels[idx] for idx in order], fontsize=12)

    # Title
    plt.title('Régression linéaire simple', fontsize=14)
    
    plt.show()


linear_regression()
```

```python
def linear_regression_and_cost_function():
    
    from sklearn.datasets import make_regression
    from sklearn.linear_model import LinearRegression

    bias = 30
    X, y, coef = make_regression(
        n_features=1, n_informative=1, bias=bias, random_state=42, coef=True, noise=40.
    )
    X_concat = np.c_[X, np.ones(X.size)]
    y /= 30
    coef_true = coef / 30
    bias_true = bias / 30

    clf_ols = LinearRegression()
    clf_ols.fit(X, y)

    coef_fitted = clf_ols.coef_.item()
    bias_fitted = clf_ols.intercept_
    
    #### OLS example ###
    
    plt.figure(figsize=(10, 4))
    
    plt.subplot(1, 2, 1)

    plt.scatter(X, y, s=10, label='Données')
    xlim = plt.xlim()
    plt.plot(xlim, [bias_fitted + xlim[0] * coef_fitted, bias_fitted + xlim[1] * coef_fitted],
             color='C1', label=fr'$y = {coef_fitted:.3f}x + {bias_fitted:.3f}$')
    plt.legend()

    plt.xlabel(r'$x$', fontsize=12)
    plt.ylabel(r'$y$', fontsize=12)
    plt.title('Régression linéaire simple')
    
    ### Fonction de coût ###
    
    # Plot OLS loss
    x_min, x_max = -2, 6

    x = np.linspace(x_min, x_max, 301)
    xx, yy = np.meshgrid(x, x)

    clf_ols.predict(np.c_[xx.ravel()])
    Z = (
        (y[None, :] - (X[:, 0] * yy.ravel()[:, None] + xx.ravel()[:, None])
    ) ** 2).mean(axis=1).reshape(xx.shape)

    plt.subplot(1, 2, 2)

    plt.imshow(Z, cmap='RdYlBu_r', origin='lower', extent=[x_min, x_max, x_min, x_max], vmin=0, vmax=30,
               alpha=0.7)
    plt.colorbar(fraction=0.0453)

    plt.contour(xx, yy, Z, colors ='k', vmin=3.1, levels=[1.5, 2, 3.6, 8, 12],
                linewidths=0.5, zorder=100)

    # Plot optimal solution for OLS
    plt.scatter(bias_fitted, coef_fitted, marker='*', s=40, color='k')

    # Plot axes
    plt.annotate(text='', xy=(0, x_max - 0.01), xytext=(0, x_min), zorder=1,
                 arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4))
    plt.annotate(text='', xy=(x_max - 0.01, 0), xytext=(x_min, 0), zorder=1,
                 arrowprops=dict(facecolor='black', shrink=0.0, width=0.1, headwidth=4, headlength=4))
    plt.text(5.8, -0.15, '$b$', ha='center', va='top');
    plt.text(-0.1, 5.8, '$a$', ha='right', va='center')
    
    plt.title('Fonction à minimiser pour trouver\nles paramètres optimaux')
    
    plt.show()

    
linear_regression_and_cost_function()
```
